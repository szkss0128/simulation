```python
#中継点とは：出口へ行けない場合に、一時的な目的地として向かう場所。
#出口が見えない時の次点の目的地
#非常口方向を示す「避難誘導標識」
#集合・退避エリア（一時避難場所）
#非常口に近い案内係（避難誘導員）
#↑エージェントからみた標識のこと


import numpy as np
import matplotlib.pyplot as plt


# セルの種類（マップ上の意味）を定数で管理
EMPTY = 0       # 通路（空白セル）
WALL = 1        # 壁（視線も遮る）
EXIT = 2        # 出口
RELAY = 3       # 中継点（標識のある場所）
COUNTER = 4     # カウンター（移動は不可・視線は通す）
SIGN = 5        # 標識（視線対象、移動不可）


def create_test_map():
    grid = np.zeros((20, 20), dtype=int)
    grid[5, 5:15] = 1
    grid[10, 10] = 1
    grid[19, 18] = 2
    grid[3, 3] = 3
    grid[12, 16] = 3
    return grid
```

```python
class Agent:
    def __init__(self, position, speed, vision_range, agent_type, mode):
        self.position = position     #現在位置（タプル）
        self.speed = speed         #速度　m/s　→ステップ調整に利用
        self.vision_range = vision_range   #視野 int（何マス先まで見えるか）
        self.agent_type = agent_type      # 若年者young, 高齢者elderly, 車イスwheelchair）
        self.mode = mode                 #動作モード（autonomous自律 or follower追従）
        self.destination = None          #現在の目的地
        self.step_counter = 0           #動けるタイミングかどうかの判定
        self.evacuated = False
        
        self.step_delay = max(1, int(round(1.4/ speed)))  #速度に応じてステップ遅延
        self.last_direction = (0, 0)  #最後に移動した方向（初期は動いていない）
```

```python
#エージェントが現在の位置から目的地に向かって1ステップ移動する。
#優先順位：直進 → 90度 → 反対方向
#移動先が空白（0）または出口（2）の場合のみ移動可

def move_towards_destination(agent, grid):
   
    if agent.destination is None:
        print(agent.agent_type,'目的地が未設定→動けません')
        return  # 目的地が未設定なら動かない

    x, y = agent.position #現在地(x, y)
    dx = agent.destination[0] - x #現在地と目的地の差をdxに代入
    dy = agent.destination[1] - y #現在地と目的地の差をdyに代入
    
    
    #1スッテプで1マスしか進まないように正規化
    step_x = np.sign(dx) #-1, 0, +1
    step_y = np.sign(dy) #-1, 0, +1


    #最優先は直進（目的地に一番近づける方向）
    directions = []
    if agent.last_direction != (0, 0):
        directions.append(agent.last_direction) #習慣的な移動
    if step_x != 0 or step_y != 0:
        directions.append((step_x, step_y)) #目的地へ1マス
    if step_x != 0 and step_y != 0:
        directions.extend([(step_x, 0), (0, step_y)]) #横・縦
    elif step_x != 0:
        directions.extend([(step_x, 1), (step_x, -1)]) #横＋上下
    elif step_y != 0:
        directions.extend([(1, step_y), (-1, step_y)]) #縦＋左右
        
    #反対方向(全てダメなとき)
    if step_x != 0 or step_y != 0:
        directions.append((-step_x, -step_y))
        
    
    #各方向を通して、通れる場所なら移動
    for d in directions:
        nx, ny = x + d[0], y + d[1] #移動候補ごとに次のマスをnx, nyとする。
        #マップ外でなければ、次のセルを確認。
        if 0 <= nx < grid.shape[0] and 0 <= ny < grid.shape[1]:
            
            #そのセルが空白(0) or 出口(2)なら、そこに移動
            if grid[nx, ny] in [EMPTY, EXIT]:
                agent.position = (nx, ny)
                agent.last_direction = d
                print(agent.agent_type,'→',agent.position,'に移動')
                return #1ステップで終了
    print(agent.agent_type,'動ける方向なし→',agent.position,'で停止')
            
            
#エージェントは目的地に向かって少しずつ進みます。
#壁(1)や他のエージェントがいるマスは避けるようにできます
#スムーズに、かつ現実的な避難経路を模擬できます。
```

```python
#×→人数が一人しか減らない

class EvacuationSimulation:
    def __init__(self, grid, agents, steps=30):
        self.grid = grid  #マップ(建物、壁、出口などを２次元配列で表現)
        self.agents = agents #Agentオブジェクトのリスト
        self.steps = steps #シミュレーションを何ステップ実行するか
        self.frames = []  # アニメーション用マップの記録
        self.remaining_counts = []  # ステップごとの残留者数
        
        
        
    #避難シミュレーション全体を動かすメイン処理run()関数
    #各エージェントはステップごとに目的地を設定して移動
    #各時間ステップのマップと残留者数を記録
    def run(self):
        from copy import deepcopy
        self.frames = []
        self.remaining_counts = []
        
        for t in range(self.steps):
            frame = self.grid.copy()
            remaining = 0
            
            for agent in self.agents:
                if agent.evacuated:
                    continue  # 既に避難済み
                    
                    
                if agent.step_counter % agent.step_delay == 0:
                    if agent.agent_type == 'young':
                        set_destination_young(agent, self.grid)
                        
                    else:
                        set_destination_elderly(agent, self.grid, self.agents)
                        
                    move_towards_destination(agent, self.grid)
                    
                agent.step_counter += 1
                
                # 出口に到達したかチェック
                if self.grid[agent.position] == EXIT:
                    agent.evacuated = True
                    continue  # 今ステップは描画・カウントしない
                    
                # エージェント描画・カウント
                x, y = agent.position
                frame[x, y] = 9
                remaining += 1
                
            self.frames.append(deepcopy(frame))
            self.remaining_counts.append(remaining)        
        
#self.frames：時間ごとのマップの変化（アニメーション用）
#self.remaining_counts：時間ごとの避難者の進行状況（グラフ用）
```

```python
#get_visible_cells()関数=視野線関数
#視野線16方向 + 壁遮蔽 + カウンターは視線通す

def get_visible_cells(grid, agent): #grid=マップ(２次元)、agent=エージェントの情報
    
    x0, y0 = agent.position #x0, y0→エージェントの現在位置
    r = agent.vision_range #r=その人の視野範囲
    visible = [] #見えるセルの結果を載せていくためのリスト

    # 16方向
    directions = [
        (-1, 0), #上（北）方向
        (-1, 1), #北東（上＋右方向）
        (0, 1), #東（右方向）
        (1, 1), #南東（下+右方向）
        (1, 0), #南（下方向）
        (1, -1), #南西（下＋左方向）
        (0, -1), #西（左方向）
        (-1, -1), #北西（上＋左方向）
        (-2, 1), #北寄りの北東（上が多め、少し右）
        (-1, 2), #東寄りの北東（右が多め、少し上）
        (1, 2), #東寄りの南東（下＋右（右が多め））
        (2, 1), #南寄りの南東（下が多め）
        (2, -1), #南寄りの南西（下＋左（下が多め））
        (1, -2), #西寄りの南西（左が多め）
        (-1, -2), #西寄りの北西（左が多め、上が少し）
        (-2, -1) #北寄りの北西（上が多め、左が少し）
    ]

    
    #重複を防ぐために既にみたセルを覚えておくset
    seen = set()
    
    
#以下のコードに修正するべき箇所あり

    
    #16の各方向に対して、順に処理を行う。
    for dx, dy in directions:
        
        #その方向に対して、１マスずつ最大rマス先まで見える範囲を伸ばす
        for step in range(1, r + 1):
            
            #ある方向にstep回進んだ先の位置(nx, ny)を計算
            nx = x0 + dx * step
            ny = y0 + dy * step
            
            #位置(nx, ny)がマップの中かどうかの確認。（外ならbreak）
            if 0 <= nx < grid.shape[0] and 0 <= ny < grid.shape[1]:
                
                #まだ、見たことのないセルなら「セル位置とセル値」をseenに追加
                pos = (nx, ny)
                if pos not in seen:
                    visible.append((pos, grid[nx, ny]))
                    seen.add(pos)

                # もし、そのセルが壁(WALL)なら、見える範囲をそこで止める
                if grid[nx, ny] == WALL:
                    break
            
            #マップの外に出たら、そこより先には進めないからbreak
            else:
                break  # マップ外に出たら終了

    #最後にみえたセルの一覧（リスト）を返す。
    return visible
```

```python
#若年者（自律型）が目的地を決定するための関数

def set_destination_young(agent, grid):
    #[ (位置、セルの種類), …]の形
    visible = get_visible_cells(grid, agent)  # 16方向視野線＋遮蔽あり

    # 優先1：出口が見えたらそこへ
    for (x, y), cell in visible:  #visibleにあるすべてのセルを調べる
        if cell == EXIT: #そのセルの値がEXITなら、
            agent.destination = (x, y) #その位置をagent.distinationとして、
            return #return（探索終了）

    # 優先2：中継点（集結場所）へ
    for (x, y), cell in visible: #visibleから中継点を探す
        if cell == RELAY: #RELAYは途中で集まる安全場所
            agent.destination = (x, y) #もしそうなら、そこへ向かう
            return #見つかったので、return（目的地が決まる）

    # 優先3：なるべく遠い空白セルへ（避難候補）
    #出口も中継点も見えなかった場合
    #visibleのセルの中から、val == EMPTY(空白)だけ選ぶ
    #→自分の現在位置との距離を計算して
    #→距離が遠い順に並べ替える
    #np.linalg.norm()はユークリッド距離（直線距離）を出す関数
    fallback = sorted(
        [(pos, np.linalg.norm( np.array(pos) - np.array(agent.position) ) )
         for pos, val in visible if val == EMPTY],
        key=lambda x: -x[1]  # 距離が遠い順に並べる
    )
    #もし、遠くの空白セルが１つ以上あれば、その中で一番遠いものを目的地
    if fallback:
        agent.destination = fallback[0][0]
    #fallback[0]=一番遠いセル
    #fallback[0][0]=その位置（タプル）だけを取り出す(x, y)
```

```python
#〇
#高齢者や車いす用の目的地関数
#agent:今動こうとしている高齢者 or 車イス
#grid: ２次元配列マップ
#agents: すべてのエージェントリスト（若年者を探すのに利用）

def set_destination_elderly(agent, grid, agents):
    visible = get_visible_cells(grid, agent)
    visible_positions = {pos for pos, _ in visible}

    # 1. 若年者が視野内にいる場合 → 一番近い人を追従
    closest = None
    min_dist = float('inf')
    for other in agents:
        if other.agent_type == 'young' and other.position in visible_positions:
            dist = np.linalg.norm(np.array(agent.position) - np.array(other.position))
            if dist < min_dist:
                closest = other.position
                min_dist = dist
    if closest:
        agent.destination = closest
        return

    # 2. 視野内に出口または標識があれば、それを目的地にする
    for (x, y), val in visible:
        if val in [EXIT, SIGN]:
            agent.destination = (x, y)
            return

    # 3. それもなければ、自律的に遠くの空白を探す（若年者と同様）
    agent.destination = None
    set_destination_young(agent, grid)
```

```python
# --- このセルはわかりません------

import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import ArtistAnimation
from IPython.display import HTML, display

# ---------- 1) 可視化用の関数 ----------
def show_animation(sim, interval=300):
    
    fig, ax = plt.subplots()
    ims = []
    for frame in sim.frames:
        im = ax.imshow(frame, animated=True, cmap='tab20')
        ims.append([im])
    ani = ArtistAnimation(fig, ims, interval=interval, repeat=False)
    plt.close(fig)           # 2重表示を防ぐ
    display(HTML(ani.to_jshtml()))

def plot_remaining(sim):
    plt.figure()
    plt.plot(sim.remaining_counts, marker='o')
    plt.xlabel("Step")
    plt.ylabel("Remaining")
    plt.grid(True)
    plt.show()

# ---------- 2) シミュレーション実行 ----------
grid = np.zeros((20, 20), dtype=int)
grid[9, 5] = EXIT            # 出口

a1 = Agent(position=(1, 5), speed=1.4, vision_range=10, agent_type="young", mode="autonomous")
a2 = Agent(position=(1, 10), speed=0.7, vision_range=10, agent_type="elderly", mode="follower")
a3 = Agent(position=(2, 2), speed=2.0, vision_range=15, agent_type="young", mode="autonomous")

agents = [a1, a2, a3]

sim = EvacuationSimulation(grid, agents, steps=130)
sim.run()

# ---------- 3) 可視化 ----------
plot_remaining(sim)   # 折れ線グラフ
show_animation(sim)    # アニメーション
```
